### Code

```python
import pandas as pd

df = pd.read_csv('Chocolate Sales.csv')
df.head()
```

```python
#data cleaning - done this a few times on the dataset so copied it over. 
df['Amount'] = df['Amount'].str.replace(r'[$,]', '', regex=True).astype(int)
df['Date'] = pd.to_datetime(df['Date'], format = '%d-%b-%y')
```

```python
#setting an index
sales_idx = df.set_index('Sales Person')
sales_idx.head()
```

```python
#loc on index
sales_idx.loc['Jehu Rudeforth'].head()
```

```python
#multi indexing by sales and product
sales_multi = df.set_index(['Sales Person', 'Product']).sort_index()
sales_multi.head()
```

```python
#subsetting multi indexes
sales_multi.loc[[('Andria Kimpton', '50% Dark Bites'), ('Jehu Rudeforth', 'Drinking Coco')]]
```
